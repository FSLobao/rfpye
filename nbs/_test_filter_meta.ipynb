```python
def spectrum_stats(df, start, stop, freq_start, freq_end):
    df = df.copy()
    try:
        start = pd.to_datetime(start)
        stop = pd.to_datetime(stop)
    except pd.errors.ParserError
    
    if 'index' in df.columns:
        #df.drop('index', inplace=True, axis=1)
        df.set_index('index', inplace=True)
        df.index = pd.to_datetime(df.index)
        
    cols = df.columns.values.astype('float')
    filter_cols = (float(freq_start) <= cols) & (cols <= float(freq_end))
    
    
    freq = np.round(cols, 4).astype('float32')
    array = df.values
    min_ = array.min(axis=0)
    max_ = array.max(axis=0)
    mean = array.mean(axis=0)
    return pd.DataFrame({'Frequency': freq, 'Min': min_, 'Max': max_, 'Mean': mean})
```
